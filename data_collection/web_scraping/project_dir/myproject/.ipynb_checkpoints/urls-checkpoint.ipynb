```python
from bs4 import BeautifulSoup as bs
import requests
import re
```

```python
url = 'https://www.rent.ie/'
```

```python
soup = bs(requests.get(url).content, 'html.parser')
soup
```

```python
soup.find('select', attrs={'id': 'county', 'name': 's[cc_id]', 'class': 'wide'}).find_all('option')
```

```python
soup.find('select', attrs={'id': 'county', 'name': 's[cc_id]', 'class': 'wide'}).find_all('option')[0].get_text()
```

```python
urls = []
for row in soup.find('select', attrs={'id': 'county', 
                                      'name': 's[cc_id]', 
                                      'class': 'wide'}).find_all('option'):
    
    city = row.get_text()
    urls.append(city)
print(len(urls))
urls
```
